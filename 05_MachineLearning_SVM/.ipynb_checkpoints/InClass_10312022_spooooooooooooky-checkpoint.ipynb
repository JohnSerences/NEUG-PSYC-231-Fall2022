# support vector machines...(pattern classification)

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

```

```python
def gen_data(num_data_pnts, mu1, mu2, plot):
    '''
    docstring goes here!!!
    '''
    
    # our response of each var to cond 1
    tmp1 = np.random.randn(num_data_pnts,1) + mu1[0]
    tmp2 = np.random.randn(num_data_pnts,1) + mu1[1]

    # our data set for condition 1
    c1_data = np.hstack( (tmp1, tmp2) )

    # our response of each var to cond 2
    tmp1 = np.random.randn(num_data_pnts,1) + mu2[0]
    tmp2 = np.random.randn(num_data_pnts,1) + mu2[1]

    # our data set for condition 1
    c2_data = np.hstack( (tmp1, tmp2) )
    
    if plot:
        
        plt.scatter(c1_data[:,0], c1_data[:,1], color='b')
        plt.scatter(c2_data[:,0], c2_data[:,1], color='r')
        plt.xlabel('var 1')
        plt.ylabel('var 2')
        plt.legend(['Cond 1', 'Cond 2'], loc='best', fontsize='x-small')
        plt.show()
        
        
    return c1_data, c2_data

    
```

```python
np.random.seed( 8 )

mu1 = [1,6]
mu2 = [6,1]

N = 100

d_c1, d_c2 = gen_data(num_data_pnts=N,mu1=mu1,mu2=mu2,plot=1)

x_range = -4,10
x_vals = np.linspace(x_range[0], x_range[1])

plt.scatter(d_c1[:,0],d_c1[:,1],color='b')
plt.scatter(d_c2[:,0],d_c2[:,1],color='r')

# new thing...
for a,b in [(.75,.65),(.7,.5),(1.15,-.1)]:
    
    plt.plot(x_vals, a * x_vals + b, 'k-')
    
plt.xlim( x_range[0], x_range[1] )

plt.plot(6,6, 'o', c='g', markersize = 10)
    
plt.show()

```

```python
# train svm (SVC)
percent_train = .9

# get the rows/columns...
r,c = d_c1.shape

trn_length = int( np.round( percent_train * r, 0) )

# train data
trn_data = np.vstack((d_c1[0:trn_length,:], d_c2[0:trn_length,:]))

trn_labels = np.hstack((np.zeros(trn_length), np.ones(trn_length)))

model = SVC(kernel='linear', C=1)

model.fit(trn_data, trn_labels)

# plot the output of the model fit...

x = np.linspace(-3, 8, 100)
y = np.linspace(-3, 10, 100)

# make combination of all possible x,y pairs...
# meshgrid...
X, Y = np.meshgrid(x,y)
print(X.shape)
XY = np.vstack(([X.ravel(), Y.ravel()])).T

DF = model.decision_function(XY).reshape(X.shape)

plt.scatter( trn_data[:,0], trn_data[:,1], c='b')

axis = plt.gca()

axis.contour(X, Y, DF, colors='k',
            levels = [-.8, 0, .8],
            linestyles = ['--', '-', '--'])

axis.scatter(model.support_vectors_[:,0],
            model.support_vectors_[:,1],
            s=100, c='g')

plt.show()
```
