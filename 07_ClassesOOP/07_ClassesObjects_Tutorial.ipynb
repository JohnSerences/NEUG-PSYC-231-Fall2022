```python
import numpy as np
import matplotlib.pyplot as plt
from AnalyzeData import *

# check these...
%load_ext autoreload
%autoreload 2
```

```python
# make our object!
# then we can get access to all functionality 
# of the object via AD
AD = AnalyzeData()
```

```python
# load our eeg data
data, sr, tx, cond_labels = AD.load_data('class_demo_eeg.npz')
```

```python
# return data after meaning across rows
mr_data = AD.mean_rows(data)

# plot the data meaned over trials (rows)
plt.plot( tx, mr_data.T )
plt.xlabel('Time(ms)')
plt.ylabel('Resp Amp')

plt.show()
```

```python
# generate a list of angles specified in degrees
# and returned as radians
rad = AD.gen_deg_rads( [0,360], 180 )

# make a sin and cos
sw = np.sin(rad)
cw = np.cos(rad)

plt.plot(rad, sw)

plt.plot(rad, cw)

plt.xlabel( 'Radians' )
plt.ylabel( 'Amplitude' )

plt.show()
```

## pass in custom params to initialize the object

```python
# file name that you want load
fn = 'class_demo_eeg.npz'
avg_type = 'rows'

# make a dictionary that contains 
params = {'file_name' : fn, 'avg_type' : avg_type}

# make the instance, pass in params
AD = AnalyzeData( params )

# load data
AD.load_data3()

# then mean and plot
AD.mean_rows2( AD.data )

# plot!
plt.plot( AD.tx, AD.mr_data.T )
plt.xlabel('Time(ms)')
plt.ylabel('Resp Amp')
plt.show()
```
